Q1. What is the Filter method in feature selection, and how does it work?

The Filter method selects features by using statistical measures to evaluate the relationship between each feature and the target variable.

It works before training any machine-learning model.

How it works (idea):

Compute a score for each feature

Rank the features

Select the top-k features

Common filter scores

Correlation

Chi-square test

ANOVA F-test

Mutual information

Exampl

```python
from sklearn.feature_selection import SelectKBest, f_classif

selector = SelectKBest(score_func=f_classif, k=5)
X_new = selector.fit_transform(X, y)
```

Q2. How does the Wrapper method differ from the Filter method?
Aspect	Filter Method	Wrapper Method
Uses ML model	❌ No	✅ Yes
Evaluation	Statistical tests	Model performance
Speed	Fast	Slow
Accuracy	Moderate	Often higher
Overfitting risk	Low	Higher

Main difference

👉 Filter methods evaluate features independently of a model,
while wrapper methods evaluate feature subsets using a trained model.


Q4. Drawbacks of the Filter method

Ignores feature interactions

Does not consider the learning algorithm

Selected features may not give the best model performance

Often selects redundant features

Q5. When would you prefer the Filter method over the Wrapper method?

You prefer the Filter method when:

The dataset has very many features

You need fast preprocessing

You want a model-independent selection

You want to reduce dimensionality before trying multiple models

Q8. House price prediction – using the Wrapper method

Here, the number of features is small and accuracy is important.

I would use a Wrapper method, such as Recursive Feature Elimination (RFE).

Steps

Choose a learning model (for example, linear regression)

Try different feature subsets

Select the subset that gives the best validation performance

Example:

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

model = LinearRegression()
rfe = RFE(model, n_features_to_select=3)

X_selected = rfe.fit_transform(X, y)
```

This method repeatedly:

fits the model

removes the least useful feature

checks performance

It is suitable here because:

the number of features is small

we want the best possible subset for this specific model
